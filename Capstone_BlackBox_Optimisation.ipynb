```python
### Imports & Configurations:

import os
import io
import base64
from datetime import datetime, timedelta
import shutil
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, Matern, RationalQuadratic, DotProduct, ExpSineSquared, WhiteKernel, ConstantKernel as C
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error
from scipy.stats import norm
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from mpl_toolkits.mplot3d import Axes3D
import json
import random
import itertools 
from joblib import Parallel, delayed
import multiprocessing 
import time # NEW: For runtime tracking
import sys # NEW: For general memory usage fallback
import uuid # NEW: To generate a unique ID for each run

# The 'resource' module is not available on Windows.
try:
    import resource
    RESOURCE_AVAILABLE = True
except ImportError:
    RESOURCE_AVAILABLE = False

import warnings
from sklearn.exceptions import ConvergenceWarning

from sklearn.neural_network import MLPRegressor

import json 
import ast  
import copy 

EXPERIMENT_LOG_FILE = "bbo_experiment_log.csv"

warnings.filterwarnings("ignore", category=ConvergenceWarning)
pd.set_option("display.max_colwidth", None)

BASE_SIZES = [0, 10, 10, 15, 30, 20, 20, 30, 40]

KERNELS_MAP = {
    # Define instances of simple kernels
    'rbf': RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2)),
    'matern': Matern(length_scale=1.0, length_scale_bounds=(1e-2, 1e2), nu=1.5),
    'rq': RationalQuadratic(length_scale=1.0, alpha=1.0),
    # Composite kernels must be treated separately in the training function
}

HISTORIC_FILE = 'historic_bbo_query_results.csv'

KERNELS_TO_TEST = ['matern', 'rbf', 'rq', 'matern+white'] # Include your stabilized kernels
ACQUISITIONS_TO_TEST = ['ucb', 'ei']

# Default configuration used for all functions unless explicitly overridden
DEFAULT_HPS = {
    "GP_FIT_N_RESTARTS": 1, 
    "UCB_BETA": 3.0,      
    "EI_XI": 0.01,        
    "SVM_C": 100,
    "NN_ALPHA": 0.0001,
    # Add keys for consistency and fallback
    "ALPHA": 1e-10, # Add default ALPHA
    "random_state": 42, # Add default random_state
    "KERNELS_TO_TEST": KERNELS_TO_TEST,
    "ACQUISITIONS_TO_TEST": ACQUISITIONS_TO_TEST,   
}

# Explicitly override defaults where needed (e.g., Function 5 might be harder to fit)

FUNCTION_HPS = {
    # CLUSTER A: Escape the flat landscape (Toned down slightly for safety)
    1: {**DEFAULT_HPS, "UCB_BETA": 5.0, "EI_XI": 0.1},
    3: {**DEFAULT_HPS, "UCB_BETA": 5.0, "EI_XI": 0.1},
    
    # CLUSTER B: The "Money" Function (Locking in the 7400+ peak)
    5: {**DEFAULT_HPS, "UCB_BETA": 1.5, "GP_FIT_N_RESTARTS": 10},
    
    # CLUSTER C+: High-D Stability (Don't lose the 3.17 breakthrough)
    7: {**DEFAULT_HPS, "UCB_BETA": 1.0, "GP_FIT_N_RESTARTS": 10},

    # CLUSTER C: Balanced (Recovery for Function 4)
    2: DEFAULT_HPS.copy(),
    4: DEFAULT_HPS.copy(), 
    6: DEFAULT_HPS.copy(),
    8: DEFAULT_HPS.copy()
}

# --- Central Hyperparameter Configuration Module ---
BBO_CONFIG = {
    # Bayesian Optimization Meta-Parameters
    'BBO_META': {
        'NUM_CANDIDATE_MULTIPLIER': 1000,
        'NORMALIZATION_METHOD': 'StandardScaler',
    },
    
    # Gaussian Process (GP) & Acquisition Specific Hyperparameters
    'GP_HPS': {
        'KERNELS_TO_TEST': ['rbf', 'matern', 'rq', 'matern+white'],
        'ALPHA': 1e-10,
        'ACQUISITIONS_TO_TEST': ['ucb', 'ei'],
        'UCB_BETA': 0.1,  # Default exploration parameter for UCB
        'EI_XI': 0.01,    # Default exploration parameter for EI
        'GP_FIT_N_RESTARTS': 10, # Keep this low for faster tuning
        'random_state': 42
    },
    
    # Neural Network (NN) Specific Hyperparameters
    'NN_HPS': {
        'hidden_layer_sizes': (64, 64),
        'activation': 'relu',
        'learning_rate_init': 0.001,
        'max_iter': 15, # Recommended for speed
        'alpha': 0.0001,
        'random_state': 42,
        # --- NEW Early Stopping Config for MLPRegressor ---
        'early_stopping': True,
        'validation_fraction': 0.2,
        'n_iter_no_change': 50, # Stop if no improvement after 50 steps
        'tol': 1e-4,
    },
    
    # SVM Filtering Specific Hyperparameters
    'SVM_HPS': {
        'SVM_C': 100,
        'SVM_KERNEL': 'rbf',
        'MEDIAN_THRESHOLD_STRATEGY': 'median',
    },
    # --- Global Parallel Config ---
    'PARALLEL_JOBS': multiprocessing.cpu_count() - 1 or 1, # Use N-1 CPU cores
}

# --- NEW: FUNCTION-SPECIFIC HYPERPARAMETER OVERRIDES ---
FUNCTION_SPECIFIC_HPS = {
    # Fix for Functions 1 & 3: High ALPHA to stop the model from fitting 
    # infinitesimal noise as "peaks". RationalQuadratic for long-range vision.
    1: {
        'GP_HPS': {
            'ALPHA': 1e-3, 
            'KERNELS_TO_TEST': ['matern+white', 'rq']
        },
        'NN_HPS': {'USE_NN': False} # NN gave negative acquisition scores
    },
    3: {
        'GP_HPS': {
            'ALPHA': 1e-4, 
            'KERNELS_TO_TEST': ['matern', 'rq']
        },
        'NN_HPS': {'USE_NN': False}
    },
    
    # Fix for Function 5: Massive ALPHA (0.2) to smooth the 7000+ spikes.
    5: {
        'GP_HPS': {
            'ALPHA': 0.2, 
            'KERNELS_TO_TEST': ['rq', 'matern'] 
        },
        'NN_HPS': {'USE_NN': False}
    },
    
    # Clustered logic for standard functions
    2: { 'GP_HPS': { 'ALPHA': 1e-5, 'KERNELS_TO_TEST': ['matern+white', 'rq'] } },
    7: { 'GP_HPS': { 'ALPHA': 1e-5, 'KERNELS_TO_TEST': ['matern+white', 'rq'] } }
}

# --- Global Logging Variable ---
BBO_LOGS = []

# --- NEW: Tuning Constants and Search Space ---
TUNING_METHODS = {
    "MANUAL": "Manual List",
    "GRID": "Grid Search",
    "RANDOM": "Random Search",
    "BAYESIAN": "Bayesian Optimization (Stub)", # Placeholder
    "HYPERBAND": "Hyperband (Stub)" # Placeholder
}

# Define the search space for your hyperparameters. The values must be lists.
# NOTE: The tuning functions will only look for parameters defined in this space.
PARAM_SPACE = {
    'GP_HPS': {
        # Test 1.0 (Exploitation), 3.0 (Default), 5.0/7.0 (Exploration)
        'UCB_BETA': [1.0, 3.0, 5.0, 7.0], 
        # Test 0.001 (Exploitation), 0.01 (Default), 0.1 (Exploration)
        'EI_XI': [0.001, 0.01, 0.1] 
    },
    'SVM_HPS': {
        # Test 100 (Default), 500 (Stricter boundary)
        'SVM_C': [100, 500] 
    },
    # Add other HPS categories here if you want to tune them (e.g., NN_HPS)
}

# Retaining original KERNELS/ACQUISITIONS variables for function compatibility
KERNELS = BBO_CONFIG['GP_HPS']['KERNELS_TO_TEST']
ACQUISITIONS = BBO_CONFIG['GP_HPS']['ACQUISITIONS_TO_TEST']

gp_kernel_config = {
    'kernel': Matern(length_scale=1.0, nu=2.5),
    'alpha': 1e-6,
    'optimizer': 'fmin_l_bfgs_b',
    'normalize_y': True
}

### Custom function to print logs for debugging

LOG_LEVEL = "NONE"  # Options: "DEBUG", "INFO", "WARN", "ERROR", "NONE"

def print_log(message, level="INFO"):
    """
    Prints a message if its level is equal to or higher than the global LOG_LEVEL.

    Parameters:
        message (str): The message to print.
        level (str): One of "DEBUG", "INFO", "WARN", "ERROR".
    """
    levels = ["DEBUG", "INFO", "WARN", "ERROR"]
    if LOG_LEVEL == "NONE":
        return
    if levels.index(level) >= levels.index(LOG_LEVEL):
        print(f"[{level}] {message}")

def get_resource_usage(start_time):
    """Calculates elapsed time and peak memory usage using a cross-platform approach."""
    global RESOURCE_AVAILABLE
    
    elapsed_time = time.time() - start_time
    max_memory_mb = 0.0
    
    if RESOURCE_AVAILABLE:
        # Detailed tracking for POSIX systems (Linux/macOS)
        try:
            # ru_maxrss is max resident set size (max memory used)
            max_rss = resource.getrusage(resource.RUSAGE_SELF).ru_maxrss
            # On Linux, ru_maxrss is in kilobytes; on macOS, it's in bytes.
            # Assuming KB (Linux common), divide by 1024 to get MB.
            # If running on macOS, the value will be too low, but it is the best we can do without checking OS type.
            max_memory_mb = max_rss / 1024
        except Exception as e:
            print_log(f"Warning: Resource usage tracking failed ({e}).", "WARN")
            
    else:
        # Fallback for systems where 'resource' isn't available (e.g., Windows)
        # This only provides the memory currently used by the Python interpreter, not the peak.
        # It's an imperfect but necessary placeholder.
        # Memory used by Python process in MB
        max_memory_mb = sys.getsizeof({}) / 1024 / 1024 
        print_log("Warning: Using basic memory tracking fallback (not peak usage).", "WARN")
        
    return elapsed_time, max_memory_mb

def log_experiment_results(log_data):
    """Appends experiment data to a global CSV log file."""
    df_new = pd.DataFrame([log_data])
    
    # Use 'a' (append) mode if file exists, else create it
    if os.path.exists(EXPERIMENT_LOG_FILE):
        df_new.to_csv(EXPERIMENT_LOG_FILE, mode='a', header=False, index=False)
    else:
        df_new.to_csv(EXPERIMENT_LOG_FILE, index=False)
    
    print_log(f"Experiment log saved to {EXPERIMENT_LOG_FILE}.", "INFO")
        
def parse_query_string(query_str):
    """Converts a hyphen-separated or string-list query into a list of floats."""
    if pd.isna(query_str):
        return []
    
    # 1. Attempt to parse as a string representation of a list 
    if isinstance(query_str, str) and (query_str.startswith('[') or query_str.startswith('array(')):
        try:
            # Handle numpy array string format
            clean_str = query_str.replace('array(', '').replace(')', '')
            return [float(x.strip()) for x in ast.literal_eval(clean_str) if x.strip()]
        except:
            pass
            
    # 2. Attempt to parse as hyphen-separated string (from CSV 'Query' column)
    try:
        return [float(x.strip()) for x in str(query_str).split('-') if x.strip()]
    except Exception:
        return []

def append_to_history(df_new_results):
    """Loads existing history, appends new results, and saves the combined DataFrame."""
    if os.path.exists(HISTORIC_FILE):
        try:
            # Use keep_default_na=False to avoid interpreting empty strings as NaN
            df_existing = pd.read_csv(HISTORIC_FILE, keep_default_na=False)
            df_combined = pd.concat([df_existing, df_new_results], ignore_index=True)
            print_log(f"Appended {len(df_new_results)} results. Total history: {len(df_combined)} rows.", "INFO")
        except Exception as e:
            print_log(f"Error reading existing history file ({e}). Overwriting with new data.", "WARN")
            df_combined = df_new_results
    else:
        df_combined = df_new_results
        print_log(f"Creating new history file with {len(df_combined)} rows.", "INFO")

    df_combined.to_csv(HISTORIC_FILE, index=False)

def get_next_run_id(function_number, week_number):
    """Calculates the next unique RunID for a given function and week based on historic data."""
    func_str = f"function {function_number}"
    
    if os.path.exists(HISTORIC_FILE):
        try:
            df_hist = pd.read_csv(HISTORIC_FILE, keep_default_na=False)
            df_filtered = df_hist[
                (df_hist['Function'] == func_str) & 
                (df_hist['Week'] == week_number)
            ].copy()
            
            if df_filtered.empty or 'RunID' not in df_filtered.columns or df_filtered['RunID'].empty:
                return 1
            
            # The max ID is the last run's ID
            max_run_id = df_filtered['RunID'].max()
            return int(max_run_id) + 1

        except Exception as e:
            print_log(f"Error reading history for RunID calculation ({e}). Assuming RunID=1.", "WARN")
            return 1
    else:
        return 1

def generate_grid_trials(param_space):
    """Generates a list of all hyperparameter combinations (trials) from the search space."""
    trials = []
    keys = []
    value_lists = []
    
    # Flatten the parameter space into parallel lists for itertools.product
    for category, hps in param_space.items():
        for key, values in hps.items():
            keys.append((category, key))
            value_lists.append(values)

    # Generate Cartesian product (all combinations)
    for combination in itertools.product(*value_lists):
        trial = {}
        for i, val in enumerate(combination):
            category, key = keys[i]
            if category not in trial:
                trial[category] = {}
            trial[category][key] = val
        trials.append(trial)
        
    return trials

def generate_random_trials(param_space, n_iter=10):
    """Generates a list of random hyperparameter combinations (trials) from the search space."""
    trials = []
    
    # Prepare flat list of parameter candidates
    keys = []
    for category, hps in param_space.items():
        for key, values in hps.items():
            keys.append((category, key, values))
            
    # Generate N random combinations
    for _ in range(n_iter):
        trial = {}
        for category, key, values in keys:
            if category not in trial:
                trial[category] = {}
            # Choose a random value from the list of candidates
            trial[category][key] = random.choice(values)
        trials.append(trial)
        
    return trials

### Load old submissions Week 0 to 2 for Week 1 to 3 respectively

# Define the function
def get_old_bbo_data(function_name):
    # Filter by function name
    filtered = prev_df[prev_df["Function"] == function_name].copy()
    
    # Sort by week
    ordered = filtered.sort_values(by="Week")
    
    return ordered.reset_index(drop=True)

### Load initial data

def load_initial_data(function_number):
    """
    Loads 'inputs.npy' and 'outputs.npy' from a specific function folder.

    Parameters:
        function_number (int): The function number (e.g., 1 for 'function_1').

    Returns:
        Tuple (inputs, outputs): Loaded NumPy arrays, or (None, None) if files are missing.
    """
    print_log(f"function_{function_number} Loading initial data")
    folder_name = f"function_{function_number}"
    folder_path = os.path.join(os.getcwd(), "initial_data", folder_name)

    input_path = os.path.join(folder_path, "initial_inputs.npy")
    output_path = os.path.join(folder_path, "initial_outputs.npy")

    try:
        inputs = np.load(input_path, allow_pickle=True)
    except Exception as e:
        print(f"Failed to load inputs from {input_path}: {e}")
        inputs = None

    try:
        outputs = np.load(output_path, allow_pickle=True)
    except Exception as e:
        print(f"Failed to load outputs from {output_path}: {e}")
        outputs = None

    # Create corresponding week array
    weeks = np.full(len(inputs), "Week_0")

    print_log(f"INPUTS : {inputs}")
    print_log(f"OUTPUTS: {outputs}")
    print_log("Initial data loaded successfully")
    return inputs, outputs, weeks

### Append weekly data

def append_weekly_until(inputs, outputs, weeks, function_number, max_week):
    """
    Appends all weekly inputs and outputs for a given function from week 0 to max_week (inclusive).

    Parameters:
        inputs (np.ndarray): Initial input array loaded from disk.
        outputs (np.ndarray): Initial output array loaded from disk.
        function_number (int): Function index (0-based).
        max_week (int): Week number up to which data should be appended (inclusive).

    Returns:
        Tuple (inputs, outputs): Updated arrays with all weekly data appended.
    """
    import numpy as np
    print_log(f"Appending weekly data for function_{function_number} from week 1..{max_week}")

    for week in range(0, max_week-1):
        try:
            print_log(f"Appending Week {week+1} for function {function_number}.")
            weekly_input = weekly_inputs[week][function_number - 1]
            weekly_output = weekly_outputs[week][function_number - 1]
        except IndexError:
            print(f"Week {week+1} or function {function_number} data missing â€” skipping.")
            continue

        # Stack inputs vertically
        inputs = np.vstack([inputs, np.array(weekly_input)])
        
        # Concatenate outputs
        outputs = np.concatenate([outputs, np.array([weekly_output], dtype=np.float64)])

        # Add corresponding week
        weeks = np.concatenate([weeks, np.array([f"Week_{week+1}"])])
        

    return inputs, outputs, weeks


### Load Data

def load_full_training_data(function_number, week_number):
    """
    Loads ALL historical queries and their outputs from previous weeks (from HISTORIC_FILE)
    to fully train the surrogate model. Falls back to original load_data if history is unavailable.
    """
    func_str = f"function {function_number}"
    X_hist = []
    y_hist = []

    # 1. Try to load ALL historical data from the CSV log
    if os.path.exists(HISTORIC_FILE):
        try:
            df_hist = pd.read_csv(HISTORIC_FILE, keep_default_na=False)
            
            # Filter for the current function, only past weeks, and only evaluated queries (Output not NaN/empty)
            df_filtered = df_hist[
                (df_hist['Function'] == func_str) & 
                (df_hist['Week'] < week_number) & # Only use data from previous weeks
                (df_hist['Output'].astype(str).str.len() > 0) # Ensure output is logged
            ].copy()

            if not df_filtered.empty:
                for query_str, output in zip(df_filtered['Query'], df_filtered['Output']):
                    query_input_list = parse_query_string(query_str) 
                    
                    if query_input_list:
                        X_hist.append(query_input_list)
                        y_hist.append(float(output))
                
                print_log(f"Trained GP/NN on {len(X_hist)} comprehensive historical queries from history file.", "INFO")
                # Fall through to step 2 to include data from the current week (loaded by original load_data)
        except Exception as e:
            print_log(f"Error loading or parsing historic data for training: {e}. Attempting fallback.", "ERROR")
    
    # 2. Add current week's data and initial data (via original function)
    try:
        inputs_current, outputs_current, _ = load_data(function_number=function_number, week_number=week_number)
        
        # Combine historical data (from file) and current data (from load_data)
        X_current = [x.tolist() for x in inputs_current]
        y_current = outputs_current.flatten().tolist()
        
        # Avoid duplicating current data if it was already in the history file (e.g. if running multiple times in a session)
        # For simplicity and robustness, we rely on the original load_data which handles initial+weekly submitted data.
        X_train_final = np.array(X_hist + X_current)
        y_train_final = np.array(y_hist + y_current).reshape(-1, 1)

        print_log(f"Total training data size (History + Current Week): {len(X_train_final)} points.", "INFO")

    except Exception as e:
        print_log(f"Fallback to load_data failed: {e}. Returning only history data.", "ERROR")
        X_train_final = np.array(X_hist)
        y_train_final = np.array(y_hist).reshape(-1, 1)
        
    return X_train_final, y_train_final

def load_data(function_number, week_number):
    """
    Loads initial inputs/outputs and appends all weekly data for the given function up to the given week.

    Parameters:
        function_number (int): Function number (1-based).
        week_number (int): Week number (0-based, inclusive).

    Returns:
        Tuple (inputs, outputs): Combined data arrays.
    """
    inputs, outputs, weeks = load_initial_data(function_number)

    if inputs is None or outputs is None:
        raise ValueError(f"Missing initial data for function_{function_number}: Inputs={inputs is None or outputs is None} Outputs={ outputs is None} ")

    inputs, outputs, weeks = append_weekly_until(inputs, outputs, weeks, function_number, week_number)

    return inputs, outputs, weeks


### Kernel Factory

# ðŸ§  Kernel Factory
def get_kernel(kernel_type):
    """
    Factory for STABILIZED Gaussian Process kernels.
    All kernels now include WhiteKernel for noise and constrained length scales.
    """
    # 1. Constrain length scale to prevent collapse to zero (Solving 'Near lower bound')
    LS_BOUNDS = (1e-2, 1e5) 
    
    # 2. Explicitly model noise for stability
    NOISE_KERNEL = WhiteKernel(noise_level=1e-3, noise_level_bounds=(1e-5, 1e1))
    
    # 3. Scale factor (ConstantKernel) allows the kernel variance to adapt
    SCALE_KERNEL = C(1.0, (1e-3, 1e3))

    k_type = kernel_type.lower()
    
    if 'matern' in k_type:
        # Matern is the workhorse for physical processes
        base = Matern(length_scale=1.0, nu=2.5, length_scale_bounds=LS_BOUNDS)
    elif 'rbf' in k_type:
        # RBF is for very smooth functions
        base = RBF(length_scale=1.0, length_scale_bounds=LS_BOUNDS)
    elif 'rq' in k_type:
        # RQ is for multi-scale functions (good for high dimensions)
        base = RationalQuadratic(length_scale=1.0, alpha=1.0, length_scale_bounds=LS_BOUNDS)
    elif 'dot' in k_type:
        return DotProduct(sigma_0=1.0) # DotProduct is usually robust on its own
    else:
        raise ValueError(f"Unsupported kernel type: {kernel_type}")
        
    # Combine: Scale * Structure + Noise
    return SCALE_KERNEL * base + NOISE_KERNEL
    
def get_adaptive_params(function_name, current_week):
    """
    Dynamic Logic Engine: Reads history and prescribes the next move.
    """
    fn_number = int(function_name.split(' ')[1])
    
    # 1. Start with the function's base HPs
    # Use .get() defensively, falling back to a copy of DEFAULT_HPS if the key is missing
    tuned_config = FUNCTION_HPS.get(fn_number, DEFAULT_HPS.copy())
    
    # 1. Load History using your existing function
    try:
        history_df = get_old_bbo_data(function_name)
    except Exception:
        print(f"[{function_name}] No history found. Using defaults.")
        return {'kernel_type': 'matern', 'ucb_beta': 2.0, 'ei_xi': 0.01}

    if history_df.empty:
        return {'kernel_type': 'matern', 'ucb_beta': 2.0, 'ei_xi': 0.01}
        
    # 2. Get recent performance (last 5 runs)
    recent = history_df.sort_values('Week', ascending=False).head(5)
    
    # 3. Analyze Stability (Did we crash or warn?)
    # Check columns for 'Parse error' or 'Near lower bound'
    # Assuming your history saves these in a 'LengthScaleWarning' or similar column
    unstable = False
    if 'LengthScaleWarning' in recent.columns:
         unstable = recent['LengthScaleWarning'].astype(str).str.contains('Near lower bound|Parse error', na=False).any()
            
    if unstable:
        print(f"[{function_name}] âš ï¸ Instability detected. Forcing MATERN kernel.")
        tuned_config['kernel_type'] = 'matern'
        tuned_config['UCB_BETA'] = 3.0
        tuned_config['EI_XI'] = 0.01
        return tuned_config

    # 4. Analyze Strategy (Are we stuck?)
    # Count how many times we 'Exploited' recently
    exploitation_count = (recent['Strategy'] == 'Exploitation').sum()
    
    # Determine best kernel from history (simple winner-takes-all)
    best_kernel = recent.sort_values('AcquisitionScore', ascending=False).iloc[0]['Kernel']

    if exploitation_count >= 4:
        tuned_config['UCB_BETA'] = 6.0 
        tuned_config['EI_XI'] = 0.2
        # ... also choose best_kernel and assign it:
        tuned_config['kernel_type'] = best_kernel 
    else:
        tuned_config['UCB_BETA'] = 2.5
        tuned_config['EI_XI'] = 0.01
        tuned_config['kernel_type'] = best_kernel

    return tuned_config


### Acquisition Functions
def acquisition_ucb(mean, std, beta): 
    return mean + beta * std

def acquisition_ei(mean, std, best_y, xi): 
    # Standard formula for Expected Improvement with xi
    epsilon = 1e-10 
    Z = (mean - best_y - xi) / (std + epsilon)
    
    # Calculate EI 
    return (mean - best_y - xi) * norm.cdf(Z) + (std + epsilon) * norm.pdf(Z)

### Strategy Classification

def classify_strategy(scores, mean, std, method):
    top_idx = np.argmax(scores)
    if method == 'ucb':
        ratio = std[top_idx] / (mean[top_idx] + 1e-8)
        return "Exploration" if ratio > 0.5 else "Exploitation"
    elif method == 'ei':
        return "Exploration" if std[top_idx] > np.median(std) else "Exploitation"
    else:
        return "Random"

### Adaptive Sampling

def get_num_candidates(dim, kernel_type):
    multiplier = {
        'matern': 1.0,
        'rbf': 2.0,
        'rq': 1.5,
        'periodic': 2.5,
        'matern+white': 1.2
    }
    BASE_CANDIDATES_SCALER = BBO_CONFIG['BBO_META']['NUM_CANDIDATE_MULTIPLIER'] 
    return int(BASE_CANDIDATES_SCALER * dim * multiplier.get(kernel_type, 1.0)) 


def train_gp_surrogate(X, y, kernel_type='rbf', config_hps=None):

    config = config_hps if config_hps is not None else DEFAULT_HPS
    # 1. Select Kernel (assuming KERNELS is globally defined)
    kernel_name = kernel_type.lower()
    alpha_value = config.get('ALPHA', BBO_CONFIG['GP_HPS']['ALPHA'])
    n_restarts = config.get('GP_FIT_N_RESTARTS', BBO_CONFIG['GP_HPS']['GP_FIT_N_RESTARTS'])
    rs = config.get('random_state', BBO_CONFIG['GP_HPS']['random_state'])
    
    if kernel_name == 'matern+white':
        # Handle the composite kernel case explicitly
        # This allows us to test a common composite structure without complexity
        kernel = Matern(length_scale=[1.0] * X.shape[1], nu=1.5) + WhiteKernel(noise_level=1e-5)
        print_log(f"Using Composite Kernel: {kernel_name}", "INFO")
    else:
        # Get the simple kernel object from the map. Fallback to Matern if name is wrong.
        kernel_base = KERNELS_MAP.get(
            kernel_name, 
            Matern(length_scale=[1.0] * X.shape[1], nu=1.5)
        )
        # Always combine with WhiteKernel for noise estimation
        kernel = WhiteKernel() + kernel_base
        
    # 2. Split data (80% train, 20% validation)
    y_1d = y.ravel()
    X_train, X_val, y_train, y_val = train_test_split(
        X, y_1d, test_size=0.2, random_state=config['random_state']
    )

    # 3. Train GP
    gp = GaussianProcessRegressor(
        kernel=kernel, 
        alpha=alpha_value,
        n_restarts_optimizer=n_restarts, 
        normalize_y=True,
        random_state=rs
    )
    
    gp.fit(X_train, y_train)

    # 4. Calculate errors (Need to pass return_std=True for GP, but MSE ignores the std)
    y_train_pred, _ = gp.predict(X_train, return_std=True)
    y_val_pred, _ = gp.predict(X_val, return_std=True)
    
    train_error = mean_squared_error(y_train, y_train_pred)
    val_error = mean_squared_error(y_val, y_val_pred)
    
    # 5. Return model, fitted kernel, and diagnostics (NEW RETURN SIGNATURE)
    return gp, gp.kernel_, train_error, val_error
### Generate Query

def generate_query(X, y, dim, acquisition_method, kernel_type,  use_nn=False, config_hps=None): 
    X = np.array(X)
    y = np.array(y)
    X_norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0) + 1e-8)

    scaler = StandardScaler()
    X_norm = scaler.fit_transform(X) # Fit and transform on the training data
    
    num_candidates = get_num_candidates(dim, kernel_type)
    candidates = np.random.uniform(0, 1, size=(num_candidates, dim))

    config = config_hps if config_hps is not None else DEFAULT_HPS
    kappa = config['UCB_BETA']
    xi = config['EI_XI']

    # 2. Generate candidates in the original unscaled domain [0, 1]
    candidates_unscaled = np.random.uniform(0, 1, size=(num_candidates, dim))
    # 3. Transform candidates using the fitted scaler
    candidates = scaler.transform(candidates_unscaled) # candidates is the scaled CANDIDATE data

    if use_nn:
        model, train_error, val_error = train_nn_surrogate(X_norm, y, config_hps=config_hps) 
        gradients = np.array([estimate_gradient(model, c) for c in candidates])
        gradient_mags = np.linalg.norm(gradients, axis=1)
        scores = -gradient_mags  # steer toward steep descent
        fitted_kernel = "MLPRegressor"
        warning_flag = ""
        strategy = "Gradient Descent"
    else:
        kernel = get_kernel(kernel_type)
        n_restarts = config_hps['GP_FIT_N_RESTARTS'] 
        gp = GaussianProcessRegressor(kernel=kernel, normalize_y=True, n_restarts_optimizer=n_restarts) 
        gp.fit(X_norm, y)
        fitted_kernel = str(gp.kernel_) 
        warning_flag = ""
        if "length_scale=" in fitted_kernel:
            try:
                # Find the index where the length_scale value starts
                ls_start_index = fitted_kernel.index("length_scale=") + len("length_scale=")
                
                # Look for the next comma or closing parenthesis to find the end of the value
                comma_index = fitted_kernel.find(",", ls_start_index)
                paren_index = fitted_kernel.find(")", ls_start_index)
                
                # Determine the true end index (the minimum of the valid terminators)
                valid_terminators = [idx for idx in [comma_index, paren_index] if idx != -1]
                
                if valid_terminators:
                    ls_end_index = min(valid_terminators)
                    # Extract the length scale part
                    ls_part = fitted_kernel[ls_start_index:ls_end_index].strip()
                else:
                    # If no terminators found, take the rest of the string
                    ls_part = fitted_kernel[ls_start_index:].strip()

                if "[" in ls_part:
                    # Handle array length scales (e.g., [0.1 0.2])
                    # Note: You can't use .split() directly on the string, 
                    # use string manipulation to extract the values
                    ls_values = [float(x) for x in ls_part.strip("[]").split()]
                    if any(ls <= 1e-5 for ls in ls_values):
                        warning_flag = "Near lower bound"
                else:
                    # Handle single float length scales (e.g., 0.228)
                    ls_value = float(ls_part)
                    if ls_value <= 1e-5:
                        warning_flag = "Near lower bound"
            except Exception:
                # Catch any unexpected formatting issues
                warning_flag = "Parse error"

        mean, std = gp.predict(candidates, return_std=True)
        
        beta = BBO_CONFIG['GP_HPS']['UCB_BETA']
        xi = BBO_CONFIG['GP_HPS']['EI_XI']

        if acquisition_method == 'ucb': 
            gp_scores = acquisition_ucb(mean, std, beta=kappa)
        elif acquisition_method == 'ei': 
            best_output = np.max(y)
            gp_scores = acquisition_ei(mean, std, best_output, xi=xi) 
        else:
            gp_scores = np.random.rand(len(mean))

        threshold = np.median(y)
        labels = (y > threshold).astype(int)
        
        C_param = BBO_CONFIG['SVM_HPS']['SVM_C']
        kernel_param = BBO_CONFIG['SVM_HPS']['SVM_KERNEL']
        
        svm_model = make_pipeline(StandardScaler(), SVC(C=C_param, kernel=kernel_param, probability=True))
        svm_model.fit(X_norm, labels.ravel())
        svm_probs = svm_model.predict_proba(candidates)[:, 1]

        scores = gp_scores * svm_probs
        strategy = classify_strategy(gp_scores, mean, std, acquisition_method)

    if use_nn:
        # CHANGE: Capture the new return values (errors)
        model, train_error, val_error = train_nn_surrogate(X, y, config_hps=config_hps)
        fitted_kernel = 'N/A'
        
        diagnostics = {
            'TrainError': train_error, 
            'ValError': val_error, 
            'FeatureLengthScales': 'N/A' # NN doesn't have simple scales
        }   
    else: # GP Surrogate
        # CHANGE: Capture the new return values
        model, fitted_kernel, train_error, val_error = train_gp_surrogate(X, y, kernel_type=kernel_type, config_hps=config_hps)       
        # GP Feature Relevance (Length Scales)
        # Length scales: large value indicates a feature is irrelevant.
        if hasattr(fitted_kernel, 'k2') and hasattr(fitted_kernel.k2, 'length_scale'):
            length_scales = fitted_kernel.k2.length_scale.tolist()
        elif hasattr(fitted_kernel, 'length_scale'):
             length_scales = fitted_kernel.length_scale.tolist()
        else:
            length_scales = 'N/A'
            
        diagnostics = {
            'TrainError': train_error, 
            'ValError': val_error, 
            'FeatureLengthScales': str(length_scales) 
        }
        
    best_idx = np.argmax(scores)
    best_point = candidates_unscaled[best_idx]
    best_score = scores[best_idx]
    denorm_point = best_point * (X.max(axis=0) - X.min(axis=0) + 1e-8) + X.min(axis=0)
    denorm_point = scaler.inverse_transform(best_point.reshape(1, -1))[0] # <-- CHANGE THIS
    return denorm_point, strategy, fitted_kernel, warning_flag, best_score, diagnostics

### Evaluate Function Data
def evaluate_function(function_number, week_value, kernel, acquisition, generate_query, use_nn=False, X_train=None, y_train=None, config_hps=None):
    """
    Evaluate a single kernel-acquisition combination for one function.

    Parameters:
    - func_number: int, number of the function to load the data and process
    - function_data: dict of {func_name: {'X': ndarray, 'y': ndarray}}
    - kernel: str, kernel type
    - acquisition: str, acquisition strategy
    - week_value: int or str, current week identifier
    - generate_query: callable with signature (X, y, dim, method, kernel_type)

    Returns:
    - result: dict with evaluation result
    """
    inputs, outputs, weeks = load_data(function_number=function_number, week_number=week_value)
    X = inputs
    y = outputs

    if X_train is None or y_train is None:
        # Fallback for plotting or testing if called outside of run_bbo
        inputs, outputs, weeks = load_data(function_number=function_number, week_number=week_value)
        X = inputs
        y = outputs
    else:
        # Use the comprehensive training data provided by run_bbo
        X = X_train
        y = y_train
        
    dim = X.shape[1]

    query, strategy, fitted_kernel, warning_flag, score, diagnostics = generate_query(
        X, y, dim, acquisition_method=acquisition, kernel_type=kernel, use_nn=use_nn, config_hps=config_hps
    )
    
    formatted = '-'.join([f'{x:.6f}' for x in query])

    return {
        'Week': week_value,
        'Function': f"function {function_number}",
        'Input': query.tolist(),  # store raw query for plotting
        'Kernel': kernel,
        'Acquisition': acquisition,
        'Query': formatted,
        'Strategy': strategy,
        'FittedKernel': fitted_kernel,
        'LengthScaleWarning': warning_flag,
        'AcquisitionScore': score,
        'TrainError': diagnostics['TrainError'],      
        'ValError': diagnostics['ValError'],          
        # Overfitting is detected when this ratio is significantly greater than 1.
        'OverfittingRatio': diagnostics['ValError'] / diagnostics['TrainError'] if diagnostics['TrainError'] > 1e-10 else 999.0, 
        'FeatureLengthScales': diagnostics['FeatureLengthScales'] # Feature relevanc
    }

### Kernel Aquisition Combos

def generate_kernel_acquisition_combos(KERNELS, ACQUISITIONS):
    """
    Generate all (kernel, acquisition) pairs.

    Returns:
    - list of (kernel, acquisition) tuples
    """
    return [(k, a) for k in KERNELS for a in ACQUISITIONS]


### Build weekly datafrome for plotting

def build_weekly_dataframe(inputs, outputs, base_size, kernel=None, acquisition=None, function_name=None):
    df = pd.DataFrame({
        'Input': [x.tolist() for x in inputs],
        'Output': outputs,
        'Week': [0 if i < base_size else (i - base_size + 1) for i in range(len(outputs))],
        'Kernel': kernel,
        'Acquisition': acquisition,
        'Function': function_name
    })
    return df



### Plot Blackbox data

def plot_blackbox_data(df, base_size=10):
    """
    Plots black-box optimization data with weekly updates, calculating Week index from base_size.

    Parameters:
    - df: DataFrame with columns ['Input', 'Output']
    - base_size: int, number of initial entries considered Week 0

    Behavior:
    - Calculates 'Week' column: 0 for base entries, then Week 1, 2, ... for each new entry
    - For 2D input: X1 vs X2 scatter, color-coded by Output, shape-coded by Week
    - For 3D input: 3D scatter (X1, X2, X3), color-coded by Output, shape-coded by Week
    - For 4Dâ€“8D input: X1 vs X2 projection, color-coded by Output, shape-coded by Week
    """
    sns.set(style='whitegrid')
    input_dim = len(df['Input'].iloc[0])
    input_cols = [f'X{i+1}' for i in range(input_dim)]
    input_df = pd.DataFrame(df['Input'].tolist(), columns=input_cols)

    # Calculate Week index
    total_size = len(df)
    week_labels = []
    for i in range(total_size):
        if i < base_size:
            week_labels.append(0)
        else:
            week_labels.append((i - base_size) + 1)

    df_expanded = pd.concat([df.drop(columns=['Input']), input_df], axis=1)
    df_expanded['Week'] = week_labels

    # Marker styles for different weeks
    week_markers = ['o', 's', '^', 'D', 'v', 'P', '*', 'X']
    unique_weeks = sorted(df_expanded['Week'].unique())
    marker_map = {week: week_markers[i % len(week_markers)] for i, week in enumerate(unique_weeks)}

    fig = plt.figure(figsize=(10, 8))

    if input_dim == 2 or input_dim >= 4:
        ax = fig.add_subplot(111)
        for week in unique_weeks:
            subset = df_expanded[df_expanded['Week'] == week]
            sc = ax.scatter(subset['X1'], subset['X2'], c=subset['Output'], cmap='viridis',
                            marker=marker_map[week], label=f'Week {week}', edgecolor='k', s=80)
        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.legend(title='Week')
        cbar = plt.colorbar(sc)
        cbar.set_label('Output')
        ax.set_title(f'Black-box Optimization Data ({input_dim}D projected to 2D)')

    elif input_dim == 3:
        ax = fig.add_subplot(111, projection='3d')
        for week in unique_weeks:
            subset = df_expanded[df_expanded['Week'] == week]
            sc = ax.scatter(subset['X1'], subset['X2'], subset['X3'], c=subset['Output'], cmap='plasma',
                            marker=marker_map[week], label=f'Week {week}', edgecolor='k', s=60)
        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.set_zlabel('X3')
        ax.legend(title='Week')
        cbar = fig.colorbar(sc, ax=ax, shrink=0.6)
        cbar.set_label('Output')
        ax.set_title('Black-box Optimization Data (3D)')

    #plt.tight_layout()
    #plt.show()
    return fig


### Strategy Evolution Plotting and Persistence ###


def load_and_combine_history(current_submissions_df):
    """
    Loads historical submissions from CSV, parses the query string, and 
    combines them with the current session's submissions.
    """
    
    # 1. Load History from CSV
    if os.path.exists(HISTORIC_FILE):
        try:
            df_hist = pd.read_csv(HISTORIC_FILE)
            # Filter for submitted rows
            df_hist = df_hist[df_hist['Submit'] == True].copy()
            
            # CRITICAL PARSING: Convert the 'Query' string into a list of floats
            df_hist['query_input'] = df_hist['Query'].apply(parse_query_string)
            
            # Use 'query_input' to ensure we have the list format for plotting
            df_hist_ready = df_hist.rename(columns={'Week': 'Week', 'Strategy': 'Strategy'})
            df_hist_ready = df_hist_ready[['Week', 'Strategy', 'query_input']].copy()
            
            print_log(f"Loaded {len(df_hist_ready)} historical submissions from {HISTORIC_FILE}", "INFO")
            
        except Exception as e:
            print_log(f"Error loading or parsing {HISTORIC_FILE} ({e}). Starting with current session's data.", "ERROR")
            df_hist_ready = pd.DataFrame()
    else:
        print_log(f"Historic file {HISTORIC_FILE} not found. Starting with current session's data.", "INFO")
        df_hist_ready = pd.DataFrame()
        
    # 2. Prepare Current Session Submissions (df_weekly_submissions)
    if current_submissions_df is not None and not current_submissions_df.empty:
        # [cite_start]The 'Input' column already holds the raw list/array from evaluate_function [cite: 45]
        df_current = current_submissions_df[current_submissions_df['Submit'] == True].copy()
        df_current = df_current.rename(columns={'Input': 'query_input'})
        df_current_ready = df_current[['Week', 'Strategy', 'query_input']].copy()
    else:
        df_current_ready = pd.DataFrame()
        
    # 3. Combine and Return
    df_combined = pd.concat([df_hist_ready, df_current_ready], ignore_index=True)
    
    # Remove rows where query_input parsing failed (empty list)
    df_combined = df_combined[df_combined['query_input'].apply(len) > 0].copy()
    
    return df_combined

def plot_strategy_evolution(function_number=None, run_id=None):
    """
    Generates plots showing the evolution and distribution of the query strategy,
    by reading the full history file and filtering for submitted rows.
    
    Includes backward compatibility for missing columns in older data.
    """
    
    if not os.path.exists(HISTORIC_FILE):
        print("Query history is empty (no file found). Run BBO iterations first.")
        return None, None

    # Load ALL historical data
    df_full_history = pd.read_csv(HISTORIC_FILE, keep_default_na=False)

    # 1. Filter for only the submitted, chosen queries.
    df_history = df_full_history[df_full_history['Submit'] == 'TRUE'].copy()

    
    if df_history.empty:
        print("No valid submitted queries found in history......")
        return None, None
        
    # 2. NEW FILTER: Filter by function number if specified
    title_suffix = ""
    if function_number is not None:
        func_str = f"function {function_number}"
        df_history = df_history[df_history['Function'] == func_str].copy()
        title_suffix += f" (Function {function_number})"
        
        if df_history.empty:
            print(f"No submitted queries found for {func_str} in history.")
            return None, None

    #3. Handle Run ID Filtering
    
    # Ensure 'RunID' column exists (backward compatibility: fills missing RunIDs with 0)
    df_history['RunID'] = df_history.get('RunID', 0).astype(int) 

    if run_id is not None:
        # Option 1: Filter by a specific run_id if provided
        df_history = df_history[df_history['RunID'] == run_id].copy()
        title_suffix += f" (Run ID {run_id} Only)"
    else:
        # Option 2: Filter for the MAX RunID (Last Submitted Run) for each Week/Function
        max_run_ids = df_history.groupby(['Week', 'Function'])['RunID'].transform('max')
        df_history = df_history[df_history['RunID'] == max_run_ids].copy()
        title_suffix += " (Last Submitted Run Only)"

    if df_history.empty:
        print("No submitted queries found after filtering.")
        return None, None
        
    # 4. Backward Compatibility for Essential Columns
    
    # A. Strategy: Use 'Random' as default if the column or the value is missing
    df_history['Strategy'] = df_history.get('Strategy', pd.Series('Random', index=df_history.index))
    #df_history['Strategy'].fillna('Random', inplace=True)
    df_history['Strategy'] = df_history['Strategy'].fillna('Random')
    
    # B. Query Input: CRITICAL PARSING
    df_history['query_input'] = df_history['Query'].apply(parse_query_string)
    
    # Remove rows where query_input parsing failed (empty list)
    df_history = df_history[df_history['query_input'].apply(len) > 0].copy()

    if df_history.empty:
        print("No valid query input data found after parsing.")
        return None, None
        
    # --- 5. Strategy Distribution Plot (Bar Chart) ---
    fig_bar, ax_bar = plt.subplots(figsize=(10, 6))
    
    # The bar chart now only shows the distribution for the filtered function(s)
    strategy_counts = df_history.groupby(['Week', 'Strategy']).size().unstack(fill_value=0)
    strategy_counts.plot(kind='bar', stacked=True, ax=ax_bar, cmap='viridis')
    
    title_suffix = f" (Function {function_number})" if function_number is not None else " (All Functions)"
    ax_bar.set_title('Query Strategy Distribution Over Weeks' + title_suffix)
    # ... (rest of bar plot remains the same) ...
    ax_bar.set_xlabel('Week')
    ax_bar.set_ylabel('Number of Queries Submitted')
    ax_bar.legend(title='Strategy')
    plt.xticks(rotation=0)
    plt.tight_layout()
    
    # --- 6. Query Point Evolution Plot (Scatter Plot) ---
    
    # Determine dimension and prepare data
    # Ensure all inputs have the same dimension (essential for plotting)
    unique_dims = df_history['query_input'].apply(len).unique()
    if len(unique_dims) > 1:
        # This should rarely happen after filtering by function, but remains as a safety warning
        print(f"Warning: Multiple input dimensions found ({unique_dims}). Plotting first dimension only.")
        input_dim = unique_dims[0]
        df_history = df_history[df_history['query_input'].apply(len) == input_dim].copy()
    else:
        input_dim = unique_dims[0]

    input_df = pd.DataFrame(df_history['query_input'].tolist(), columns=[f'X{i+1}' for i in range(input_dim)])
    df_history_expanded = pd.concat([df_history.reset_index(drop=True), input_df], axis=1)

    fig_scatter = plt.figure(figsize=(10, 8))
    
    strategy_colors = {'Exploration': 'tab:blue', 'Exploitation': 'tab:red', 'Gradient Descent': 'tab:green', 'Random': 'tab:grey'}
    
    if input_dim == 2 or input_dim >= 4:
        ax_scatter = fig_scatter.add_subplot(111)
        for strategy, color in strategy_colors.items():
            subset = df_history_expanded[df_history_expanded['Strategy'] == strategy]
            if not subset.empty:
                ax_scatter.scatter(subset['X1'], subset['X2'], c=color, label=strategy, marker='o', edgecolor='k', s=100, alpha=0.7)
        ax_scatter.set_xlabel('X1')
        ax_scatter.set_ylabel('X2')
        ax_scatter.legend(title='Strategy')
        ax_scatter.set_title(f'Chosen Query Locations by Strategy ({input_dim}D projected)' + title_suffix)

    elif input_dim == 3:
        ax_scatter = fig_scatter.add_subplot(111, projection='3d')
        for strategy, color in strategy_colors.items():
            subset = df_history_expanded[df_history_expanded['Strategy'] == strategy]
            if not subset.empty:
                ax_scatter.scatter(subset['X1'], subset['X2'], subset['X3'], c=color, label=strategy, marker='o', edgecolor='k', s=80, alpha=0.7)
        ax_scatter.set_xlabel('X1')
        ax_scatter.set_ylabel('X2')
        ax_scatter.set_zlabel('X3')
        ax_scatter.legend(title='Strategy')
        ax_scatter.set_title('Chosen Query Locations by Strategy (3D)' + title_suffix)
    else:
        return fig_bar, None 

    plt.tight_layout()
    
    return fig_bar, fig_scatter

### Run Blackbox optimisation

def run_bbo(function_number, week_number, base_size=10, kernel='rbf', acquisition='ei', use_nn=False, config_hps=None):     

    X_train, y_train = load_full_training_data(function_number, week_number) # <-- NEW LINE
    
    inputs, outputs, weeks = load_data(function_number, week_number)
    
    # For one combo
    result = evaluate_function(
        function_number=function_number,
        week_value=week_number,
        kernel=kernel,
        acquisition=acquisition,
        generate_query=generate_query,
        use_nn=use_nn,
        X_train=X_train, 
        y_train=y_train,
        config_hps=config_hps
    )

    bbo_result = pd.DataFrame([result])
    return bbo_result    

def plot_results(function_number, week_number, base_size=10, kernel='rbf', acquisition='ei'):
    inputs, outputs, weeks = load_data(function_number, week_number)
    df_plot = build_weekly_dataframe(inputs, outputs, base_size, kernel, acquisition, function_name=f"function_{function_number}")
    fig = plot_blackbox_data(df_plot, base_size)
    pdf = pd.DataFrame(list(zip(inputs, outputs, weeks)), columns=["Input", "Output", "Week"])


    # Convert plot to base64 image
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight')
    buf.seek(0)
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    plt.close(fig)

    # Display side by side using HTML
    html = "<style>" \
           ".output_wrapper, .output {" \
           "    overflow: visible !important;" \
           "    max-height: none !important;" \
           "}" \
           "</style>"
    
    html += "<div style='display: flex; align-items: flex-start; width: 100%; overflow: visible;'>"
    html += "<div style='flex: 1; padding-right: 20px;'>"
    html += "<img src='data:image/png;base64," + img_base64 + "' style='max-width: 100%; height: 1000;'/>"
    html += "</div>"
    html += "<div style='flex: 1;'>"
    html += pdf.to_html(index=False)
    html += "</div></div>"
    display(HTML(html))

### Evaluate function for combinations of kernels and aquisitions

def run_all_combos_bbo(function_number, week_number, base_size=10, kernels=None, acquisitions=None, config_hps=None):
    config_to_use = config_hps if config_hps is not None else BBO_CONFIG

    if kernels is None:
        kernels = config_to_use['KERNELS_TO_TEST'] 
    if acquisitions is None:
        acquisitions = config_to_use['ACQUISITIONS_TO_TEST']

    kappa = config_hps['UCB_BETA']
    xi = config_hps['EI_XI']

    results = []
    for use_nn in [False]: #[False]
        for kernel in kernels:
            for acquisition in acquisitions:
                df = run_bbo(
                    function_number=function_number,
                    week_number=week_number,
                    base_size=base_size,
                    kernel=kernel,
                    acquisition=acquisition,
                    use_nn=use_nn,
                    config_hps=config_to_use
                )
                df['Surrogate'] = 'NN' if use_nn else 'GP'
                results.append(df)

    df_results = pd.concat(results, ignore_index=True)
    best_idx = df_results['AcquisitionScore'].idxmax()
    df_results['Submit'] = False
    df_results.loc[best_idx, 'Submit'] = True

    current_run_id = get_next_run_id(function_number, week_number)
    df_results['RunID'] = current_run_id # <--- NEW COLUMN ASSIGNED TO ALL ROWS

    # NEW: Use the config_to_use dictionary
    current_config = { 
        'UCB_BETA': config_to_use['UCB_BETA'],
        'EI_XI': config_to_use['EI_XI'],
        'ALPHA': config_to_use['ALPHA'], # Add ALPHA if you are tuning it
        'GP_FIT_N_RESTARTS': BBO_CONFIG['GP_HPS']['GP_FIT_N_RESTARTS'],
        'SVM_C': BBO_CONFIG['SVM_HPS']['SVM_C'],
        'NN_ALPHA': BBO_CONFIG['NN_HPS']['alpha'],
    }

    config_json = json.dumps(current_config) 
    df_results['ConfigParameters'] = config_json

    append_to_history(df_results)
    global df_weekly_submissions
    if df_weekly_submissions is None:
        df_weekly_submissions = df_results[df_results["Submit"]].copy()
    else:
        df_weekly_submissions = pd.concat(
            [df_weekly_submissions, df_results[df_results["Submit"]]],
            ignore_index=True
        )

    return df_results

# Week 4
from sklearn.neural_network import MLPRegressor

# ... (Your existing imports around Week 4) ...

def train_nn_surrogate(X, y, config_hps=None):
    config = config_hps if config_hps is not None else DEFAULT_HPS
    y_1d = y.ravel()
    nn_alpha = config.get('NN_ALPHA', BBO_CONFIG['NN_HPS']['alpha'])
    nn_config = copy.deepcopy(BBO_CONFIG['NN_HPS'])
    nn_config['alpha'] = nn_alpha # <<< Override alpha with the function-specific value
    
    model = MLPRegressor(
            hidden_layer_sizes=nn_config['hidden_layer_sizes'],
            activation=nn_config['activation'],
            learning_rate_init=nn_config['learning_rate_init'],
            max_iter=nn_config['max_iter'],
            alpha=nn_config['alpha'],
            random_state=nn_config['random_state'],
            early_stopping=nn_config['early_stopping'],
            validation_fraction=nn_config['validation_fraction'],
            n_iter_no_change=nn_config['n_iter_no_change'],
            tol=nn_config['tol']
        )
    
    model.fit(X, y_1d) # Model internally splits X, y_1d for validation

    # Diagnostics Calculation: Manually split the data for a consistent MSE report
    X_train, X_val, y_train, y_val = train_test_split(
        X, y_1d, test_size=nn_config['validation_fraction'], random_state=nn_config['random_state']
    )
    
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)
    
    train_error = mean_squared_error(y_train, y_train_pred)
    val_error = mean_squared_error(y_val, y_val_pred)
    
    return model, train_error, val_error


def estimate_gradient(model, x, epsilon=1e-4):
    grad = np.zeros_like(x)
    for i in range(len(x)):
        x_up = x.copy()
        x_down = x.copy()
        x_up[i] += epsilon
        x_down[i] -= epsilon
        grad[i] = (model.predict([x_up])[0] - model.predict([x_down])[0]) / (2 * epsilon)
    return grad

def _run_single_trial_parallel(function_number, week_number, base_size, kernel, acquisition, use_nn, config_override, tuning_method):
    """
    Wrapper function to run a single combination, designed for parallel execution.
    It temporarily sets the BBO_CONFIG for this process thread.
    """
    global BBO_CONFIG
    import copy # Required for clean state management

    # --- NEW: Load data and check size (ROBUSTNESS CHECK) ---
    try:
        # load_full_training_data must be available globally or imported inside the function
        # Calling it here is redundant but ensures we don't proceed with insufficient data
        X_train, y_train = load_full_training_data(function_number, week_number)

        MIN_SAMPLES = 4 # Minimum points required for stable GP/NN training
        if len(X_train) < MIN_SAMPLES:
            print_log(f"Skipping trial due to insufficient data ({len(X_train)} samples). Min required: {MIN_SAMPLES}.", "WARN")
            BBO_CONFIG = original_config # CRITICAL: Restore config before returning None
            return None 

    except Exception as e:
        print_log(f"Parallel data loading failed: {e}", "ERROR")
        BBO_CONFIG = original_config # CRITICAL: Restore config before returning None
        return None
        
    # 1. Temporarily apply the trial-specific hyperparameters
    original_config = copy.deepcopy(BBO_CONFIG)
    temp_config = copy.deepcopy(original_config)
    
    for category, hps in config_override.items():
        if category in temp_config:
            temp_config[category].update(hps)
        else:
            temp_config[category] = hps
    
    # Set the global config for this thread
    BBO_CONFIG = temp_config 
    
    # 2. Execute the BBO run using the temporary config
    try:
        # run_bbo will internally call load_full_training_data and evaluate_function
        df = run_bbo(
            function_number=function_number,
            week_number=week_number,
            base_size=base_size,
            kernel=kernel,
            acquisition=acquisition,
            use_nn=use_nn
        )
        # Add the hyperparameter config used for later grouping/comparison
        config_json = json.dumps(config_override)
        df['ConfigParameters'] = config_json 
        df['Surrogate'] = 'NN' if use_nn else 'GP'
        df['TuningMethod'] = tuning_method # <-- NEW: Log the method source
        
    except Exception as e:
        print_log(f"Parallel trial failed ({kernel}/{acquisition}/{'NN' if use_nn else 'GP'}): {e}", "ERROR")
        df = None

    # 3. CRITICAL: Restore the original global configuration
    BBO_CONFIG = original_config
    
    # Return the full DataFrame row
    return df
    
def _execute_tuning_trials(function_number, week_number, base_size, experiment_list, tuning_method):
    """
    INTERNAL: Executes a list of hyperparameter tuning trials using joblib for parallel execution.
    """
    global BBO_CONFIG
    import copy 
    original_config = copy.deepcopy(BBO_CONFIG)
    
    # 1. Generate all unique parameter sets (the outer loop)
    # The inner loop (kernel/acquisition/surrogate) will be handled by the parallel wrapper
    unique_config_overrides = [json.dumps(c) for c in experiment_list]
    
    # We must ensure we run *all* combinations (kernels, acquisitions, surrogates) 
    # for *each* hyperparameter override.

    trial_params = []
    for config_override in experiment_list:
        for use_nn in [False, True]:
            for kernel in KERNELS:
                for acquisition in ACQUISITIONS:
                    trial_params.append({
                        'config_override': config_override,
                        'kernel': kernel,
                        'acquisition': acquisition,
                        'use_nn': use_nn,
                        'tuning_method': tuning_method
                    })

    # 2. Execute in Parallel (Core change)
    print_log(f"Executing {len(trial_params)} combinations across {BBO_CONFIG['PARALLEL_JOBS']} cores...", "INFO")

    # Use joblib to parallelize the execution of the _run_single_trial_parallel wrapper
    parallel_results_list = Parallel(n_jobs=BBO_CONFIG['PARALLEL_JOBS'], verbose=0)(
        delayed(_run_single_trial_parallel)(
            function_number=function_number,
            week_number=week_number,
            base_size=base_size,
            kernel=p['kernel'],
            acquisition=p['acquisition'],
            use_nn=p['use_nn'],
            config_override=p['config_override'],
            tuning_method=p['tuning_method']
        )
        for p in trial_params
    )
    
    # 3. Process Parallel Results
    
    # Restore the original config immediately
    BBO_CONFIG = original_config

    valid_results = [df for df in parallel_results_list if df is not None]

    if not valid_results:
        print_log("Parallel execution failed: No valid results were returned by any job.", "ERROR")
        # Ensure the final BBO run uses the default config if all tuning fails
        return {}, -np.inf
    
    # Concatenate all non-None DataFrames from the parallel results
    all_results_df = pd.concat(valid_results, ignore_index=True)

    if all_results_df.empty:
        print_log("Parallel execution failed to yield any results.", "ERROR")
        return {}, -np.inf

    # 4. Find the globally best result
    
    # Group by the set of ConfigParameters (which uniquely defines the outer trial)
    best_results_by_config = all_results_df.loc[all_results_df.groupby('ConfigParameters')['AcquisitionScore'].idxmax()]
    
    # Find the globally best result across all trials
    best_row_global = best_results_by_config.loc[best_results_by_config['AcquisitionScore'].idxmax()]
    
    best_overall_score = best_row_global['AcquisitionScore']
    winning_method = best_row_global['TuningMethod']
    
    try:
        # Convert JSON string back to dict
        best_config_found = json.loads(best_row_global['ConfigParameters'])
    except Exception:
        best_config_found = {}

    # 5. Log All Results from Parallel Run
    # CRITICAL: Since run_all_combos_bbo is no longer called sequentially, we log all results here
    append_to_history(all_results_df)

    # Final logging and return
    print_log("\nAll parallel tuning trials complete. Restored original BBO_CONFIG.", "HEADER")
    
    print_log(f"\n--- Best Configuration Among All Trials ---", "HEADER")
    print_log(f"Best Max Acquisition Score: {best_overall_score:.4f}", "INFO")
    print_log(f"Winning Overrides: {best_config_found}", "INFO")
    
    return best_config_found, best_overall_score, winning_method

def tune_hyperparameters(method, function_number, week_number, base_size, 
                         param_space=PARAM_SPACE, manual_trials=None, n_iter=10):
    """
    Selects and executes the specified hyperparameter tuning method.

    Args:
        method (str): The tuning method to use ("GRID", "RANDOM", "MANUAL", etc.).
        ... (other args) ...
        
    Returns:
        dict: The best hyperparameter configuration found.
    """
    
    method = method.upper()
    print_log(f"Selected Tuning Method: {TUNING_METHODS.get(method, method)}", "HEADER")
    
    trials_to_run = []

    if method == "GRID":
        trials_to_run = generate_grid_trials(param_space)
        print_log(f"Generated {len(trials_to_run)} trials for exhaustive Grid Search.", "INFO")
        
    elif method == "RANDOM":
        trials_to_run = generate_random_trials(param_space, n_iter)
        print_log(f"Generated {len(trials_to_run)} trials for Random Search (n_iter={n_iter}).", "INFO")
        
    elif method == "MANUAL":
        if not manual_trials:
            print_log("MANUAL method selected, but 'manual_trials' list is empty. No tuning performed.", "WARN")
            return {} 
        trials_to_run = manual_trials
        print_log(f"Running {len(trials_to_run)} trials defined manually.", "INFO")
        
    elif method in ["BAYESIAN", "HYPERBAND"]:
        print_log(f"{method} tuning is not yet implemented. Falling back to Grid Search.", "WARN")
        trials_to_run = generate_grid_trials(param_space)

    else:
        print_log(f"Unknown tuning method '{method}'. Running with default config.", "ERROR")
        return {}
        
    if not trials_to_run:
         print_log("No trials to run. Returning current BBO_CONFIG settings.", "INFO")
         return {}

    best_overrides, best_score, winning_method = _execute_tuning_trials( 
                function_number=function_number, 
                week_number=week_number, 
                base_size=base_size, 
                experiment_list=trials_to_run,
                tuning_method=method)

    return best_overrides, best_score, winning_method

def run_all_tuning_methods(method_list, function_number, week_number, base_size, **kwargs):
    """
    Runs multiple tuning methods sequentially and returns the best configuration found across all of them.
    """
    best_config_global = {}
    best_score_global = -np.inf
    winning_method = "N/A"
    winning_tuning_method = "DEFAULT"
    
    print_log("--------------------------------------------------", "HEADER")
    print_log("Starting SEQUENTIAL HYPERPARAMETER TUNING", "HEADER")
    print_log("--------------------------------------------------", "HEADER")
    
    for method in method_list:
        print_log(f"\n======== Executing Method: {method} ========", "HEADER")
        
        # Execute the tuning method
        best_config_method, best_score_method, method_winner = tune_hyperparameters(
            method=method, 
            function_number=function_number, 
            week_number=week_number, 
            base_size=base_size, 
            **kwargs # Pass parameters like manual_trials and n_iter
        )
        
        if best_score_method > best_score_global:
            best_score_global = best_score_method
            best_config_global = best_config_method
            winning_method = method
            winning_tuning_method = method_winner
            print_log(f"New global best score found by {method}: {best_score_global:.4f}", "SUCCESS")
        
        print_log(f"Completed {method}. Best score for this method: {best_score_method:.4f}", "INFO")
            
    print_log(f"\n==============================================", "HEADER")
    print_log(f"GLOBAL WINNER: {winning_method} with Score: {best_score_global:.4f}", "SUCCESS")
    print_log(f"Winning HPs: {best_config_global}", "SUCCESS")
    print_log(f"==============================================", "HEADER")
    
    return best_config_global, winning_tuning_method

tuning_trials = [
    # Trial 1: Baseline (default values for comparison, by providing an empty override)
    {}, 

    # Trial 2: High Exploration UCB (Higher beta emphasizes uncertainty)
    {'GP_HPS': { 'UCB_BETA': 5.0, 'EI_XI': 0.01 }}, 
    
    # Trial 3: High Exploitation UCB (Lower beta emphasizes predicted mean)
    {'GP_HPS': { 'UCB_BETA': 1.0, 'EI_XI': 0.01 }}, 

    # Trial 4: High Exploitation EI (Lower xi makes finding an improvement easier)
    {'GP_HPS': { 'UCB_BETA': 3.0, 'EI_XI': 0.001 }}, 
    
    # Trial 5: High Exploration EI (Higher xi requires a larger, more certain improvement)
    {'GP_HPS': { 'UCB_BETA': 3.0, 'EI_XI': 0.1 }}, 
    
    # Trial 6: Stricter SVM Filter (Higher C for tighter boundary on candidate selection)
    {'SVM_HPS': { 'SVM_C': 500 }}
]

def run_experiments(function_number):
    fn = function_number 

    print_log(f"Starting automated tuning process for Function {fn}, Week {wk}.", "HEADER")
    
    updated_history_df = run_tuning_experiments(
        function_number=fn, 
        week_number=wk, 
        base_size=BASE_SIZES[fn], 
        experiment_list=tuning_trials # Pass the list defined above
    )

    # Example: Plot the last (best) run for this function
    fig_bar, fig_scatter = plot_strategy_evolution(function_number=fn) 

    return pd.DataFrame(BBO_LOGS)

def run_adaptive_bbo_loop(function_number, num_iterations=10):
    """
    Runs the adaptive loop 10 times for immediate stabilization and tuning.
    """
    func_name = f"function {function_number}"
    all_results = []
    print(f"\n--- Starting Adaptive Run for {func_name} ---")
    start_time = time.time()
    for i in range(num_iterations):
        print(f"\n>> Iteration {i+1}/{num_iterations}")
        it_start_time = time.time()
        # 1. Ask the Adaptive Manager what to do
        tuned_config = get_adaptive_params(func_name, wk)
        
        # 2. Run the BBO function, passing the specific configuration
        # IMPORTANT: run_all_combos_bbo must now accept this 'config' argument
        df_results = run_all_combos_bbo(
            function_number=function_number,
            week_number=wk,
            base_size=BASE_SIZES[function_number],
            config_hps=tuned_config # Pass the specific configuration
        )
        df_results['Iteration'] = i+1
        all_results.append(df_results)
        it_end_time = time.time()
        it_elapsed_time = it_end_time - it_start_time
        print(f"Iteration {i+1}/{num_iterations} elapsed time:", str(timedelta(seconds=int(it_elapsed_time))))
        
        # 4. Save results (your code already does this in run_all_combos_bbo)
        # The loop naturally feeds back into itself because get_adaptive_params reads the file!

    df_combined = pd.concat(all_results, ignore_index=True)
    #print(df_combined)
    best_row_idx = df_combined['AcquisitionScore'].idxmax()
    df_combined['Submit'] = False
    df_combined.loc[best_row_idx, 'Submit'] = True
    #print(df_combined)

    end_time = time.time()

    # Convert to hh:mm:ss
    elapsed_time = end_time - start_time
    print("Elapsed time:", str(timedelta(seconds=int(elapsed_time))))

    global BBO_LOGS

    # Find the row that was marked Submit=True (the best result)
    best_row = df_combined[df_combined["Submit"]].iloc[0]
    
    log_entry = {
        'week': wk,
        'function_number': function_number,
        'surrogate': best_row['Surrogate'],
        'kernel': best_row['Kernel'],
        'acquisition': best_row['Acquisition'],
        'strategy': best_row['Strategy'],
        'acquisition_score': best_row['AcquisitionScore'],
        'chosen_query': best_row['Query'],
        'tuning_parameters_used': {
            'UCB_BETA': BBO_CONFIG['GP_HPS']['UCB_BETA'],
            'EI_XI': BBO_CONFIG['GP_HPS']['EI_XI'],
            'GP_FIT_N_RESTARTS': BBO_CONFIG['GP_HPS']['GP_FIT_N_RESTARTS'],
            'SVM_C': BBO_CONFIG['SVM_HPS']['SVM_C'],
            'NN_ALPHA': BBO_CONFIG['NN_HPS']['alpha'],
        },
        'justification': f"Chosen because it yielded the maximum acquisition score ({best_row['AcquisitionScore']:.4f}) out of all {len(df_results)} combinations tested for function {function_number} at week {wk}.",
    }
    BBO_LOGS.append(log_entry)
    
    return df_combined

def execute_bbo_pipeline(function_number):
    """
    Executes the entire BBO workflow for a given function: 
    1. Tunes hyperparameters using multiple methods.
    2. Applies the best configuration found.
    3. Runs the final submission BBO process.
    4. Plots the results.
    
    Args:
        function_number (int): The ID of the function (1 through 8) to process.
    """
    
    # Constants required from global scope
    # NOTE: Assuming wk and BASE_SIZES are defined globally at the top of bbo.txt
    global wk
    global BBO_CONFIG
    
    fn = function_number
    if fn < 1 or fn >= len(BASE_SIZES):
        print_log(f"Error: Function number {fn} is out of valid range (1 to {len(BASE_SIZES) - 1}).", "ERROR")
        return

    base_size = BASE_SIZES[fn]
    
    print_log(f"==================================================", "HEADER")
    print_log(f"STARTING FULL BBO PIPELINE FOR FUNCTION {fn} (Week {wk})", "HEADER")
    print_log(f"==================================================", "HEADER")

    start_time = time.time()
    pipeline_run_id = str(uuid.uuid4()) # Generate a unique ID for this entire pipeline run
    print_log(f"Pipeline Run ID: {pipeline_run_id}", "HEADER")

    original_gp_config = BBO_CONFIG['GP_HPS'].copy() 
    
    if function_number in FUNCTION_SPECIFIC_HPS and FUNCTION_SPECIFIC_HPS[function_number]:
        print_log(f"Applying function-specific HPs for Function {function_number}.", "INFO")
        function_hps = FUNCTION_SPECIFIC_HPS[function_number]
        
        # Merge the specific HPs into the global BBO_CONFIG
        for category, hps in function_hps.items():
            if category in BBO_CONFIG:
                BBO_CONFIG[category].update(hps)
                
    # --- 1. Define Tuning Parameters (Moved into the pipeline) ---
    
    # Sequence of tuning methods to run
    TUNING_METHODS_TO_RUN = ["RANDOM", "MANUAL"] 

    # Manual trials (used when "MANUAL" is in the list above)
    manual_trials_list = [
        # Test a high exploration beta
        {'GP_HPS': { 'UCB_BETA': 6.0 }}, 
        # Test a high exploitation xi
        {'GP_HPS': { 'EI_XI': 0.0005 }}
    ]

    # --- 2. Execute Sequential Tuning ---
    best_overrides, winning_tuning_method = run_all_tuning_methods(
        method_list=TUNING_METHODS_TO_RUN,
        function_number=fn, 
        week_number=wk, 
        base_size=base_size, 
        manual_trials=manual_trials_list,
        n_iter=15 # Set number of iterations for Random Search
    )

    # --- 3. Apply the Best Configuration ---
    if best_overrides:
        print_log("Applying global best hyperparameters found to BBO_CONFIG for final submission run...", "SUCCESS")
        for category, hps in best_overrides.items():
            if category in BBO_CONFIG:
                BBO_CONFIG[category].update(hps)
            
    # --- 4. Run the Final BBO Selection ---
    # This run uses the best-found parameters and saves the final submission query
    print_log("\n--- Final BBO Run with Tuned Parameters ---", "HEADER")
    df_submission = run_all_combos_bbo(function_number=fn, week_number=wk, base_size=base_size)
    
    # --- 5. Log and Plot Results ---
    print_log(f"Final chosen query (using tuned HPs):\n{df_submission[df_submission['Submit']]}", "SUCCESS")
    
    fig_bar, fig_scatter = plot_strategy_evolution(function_number=fn)
    print_log(f"Generated strategy evolution plots for Function {fn}.", "INFO")

    elapsed_time, max_memory_mb = get_resource_usage(start_time)

    if not df_submission.empty and 'Submit' in df_submission.columns:
        best_row_for_submission = df_submission[df_submission['Submit']].iloc[0]
        
        experiment_log_data = {
            'PipelineRunID': pipeline_run_id,
            'FunctionNumber': fn,
            'Week': wk,
            'TuningMethodsUsed': str(TUNING_METHODS_TO_RUN),
            'WinningTuningMethod': winning_tuning_method,
            'FinalSurrogate': best_row_for_submission['Surrogate'],
            'FinalKernel': best_row_for_submission['Kernel'],
            'FinalAcquisition': best_row_for_submission['Acquisition'],
            'FinalScore': best_row_for_submission['AcquisitionScore'],
            'Time_Elapsed_seconds': elapsed_time,
            'Memory_Peak_MB': max_memory_mb,
            'BestHPsApplied': str(best_overrides),
            'Assumptions': 'Data is normalized (0-1). WhiteKernel is used for GP to model noise. Final model chosen by max AcquisitionScore.',
            'Date': pd.Timestamp.now().isoformat()
        }
        log_experiment_results(experiment_log_data)
    else:
        print_log("Warning: Submission DataFrame was empty. Skipping detailed experiment logging.", "WARN")

    BBO_CONFIG['GP_HPS'] = original_gp_config
    # Optionally, return the submission DataFrame
    return df_submission

```
